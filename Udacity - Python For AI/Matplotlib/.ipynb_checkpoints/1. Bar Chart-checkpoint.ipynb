## Qualitative Variable

Qualitative variables, also known as categorical variables, are variables that represent qualities or characteristics rather than numerical values. These variables can be divided into distinct categories or groups. Examples of qualitative variables include gender (male or female), color (red, blue, green), and education level (high school, college, graduate).

Unlike quantitative variables that can be measured and assigned numerical values, qualitative variables cannot be expressed in numerical terms. Instead, they are typically represented by labels or names that describe the different categories or groups.

Qualitative variables are often used to describe characteristics or attributes of individuals or objects in a dataset. They are commonly used in surveys, market research, and social sciences to gather information about people's preferences, opinions, or demographic characteristics.

For example, if you conducted a survey to collect data on people's favorite fruits, the variable "favorite fruit" would be a qualitative variable. The categories or groups in this case would be the different types of fruits, such as apples, oranges, and bananas.


## Quantitative Variable 

Quantitative variables, also known as numerical variables, are variables that represent numerical values or measurements. These variables can take on a range of numerical values and can be further classified as either discrete or continuous.

Discrete quantitative variables are variables that can only take on specific, separate values. For example, the number of siblings a person has or the number of cars in a parking lot are discrete variables because they can only be whole numbers.

On the other hand, continuous quantitative variables are variables that can take on any value within a certain range. Examples of continuous variables include age, height, weight, and temperature. These variables can have decimal values and can be measured with precision.

Quantitative variables are often used in scientific experiments, statistical analysis, and financial modeling. They allow for mathematical operations such as addition, subtraction, multiplication, and division to be performed on the data.

For example, if you were conducting a study on the heights of students in a class, the variable "height" would be a quantitative variable. The heights could be measured in inches or centimeters and would represent numerical values.

In summary, quantitative variables are variables that represent numerical values or measurements. They can be discrete or continuous and are commonly used in scientific and statistical analysis.

## Bar Chart Using Seaborn

## Example 1. Create a vertical bar chart using Seaborn, with default colors

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline 
```

```python
# Read the csv file, and check its top 10 rows
pokemon = pd.read_csv('./pokemon.csv')
```

```python
pokemon.head(10)
```

```python
sb.countplot(data=pokemon, x='generation_id');
```

## Example 2. Create a vertical bar chart using Seaborn, with a uniform single color

```python
base_color = sb.color_palette()
```

```python
base_color
```

```python
base_color = sb.color_palette()[3]
```

```python
sb.countplot(data=pokemon, x='generation_id', color=base_color)
```

## Bar Chart using the Matplotlib

## Example 3. Create a vertical bar chart using Matplotlib, with a uniform single color

```python
x = pokemon['generation_id'].unique()
```

```python
y = pokemon['generation_id'].value_counts(sort=False)
```

```python
plt.bar(x,y)
```

```python
plt.bar(x,y)

#labeling the axes 
plt.xlabel('generation_id')
plt.ylabel('count')

#display the plot
plt.show()
```

## Example 4. Static and dynamic ordering of the bars in a bar chart using seaborn.countplot()

For nominal-type data, one common operation is to sort the data in terms of frequency. In the examples shown above, you can even order the bars as desirable. With our data in a pandas DataFrame, we can use various DataFrame methods to compute and extract an ordering, then set that ordering on the "order" parameter:

This can be done by using the order argument of the countplot() function.

```python
# Static-ordering the bars 
sb.countplot(data=pokemon, x='generation_id', order=[5, 1, 3, 4, 2, 7, 6] )
```

### Dynamic-ordering the bars

```python
freq = pokemon['generation_id'].value_counts()
```

```python
freq
```

```python
gen_order = freq.index
```

```python
gen_order
```

```python
sb.countplot(data=pokemon, x='generation_id', order=gen_order)
```

## Example 5. Rotate the category labels (not axes)

The category labels in the examples above are very small. In case, the category labels have large names, you can make use of the matplotlib.pyplot.xticks(rotation=90) function, which will rotate the category labels (not axes) counter-clockwise 90 degrees

```python
sb.countplot(data=pokemon, x='type_1')
plt.xticks(rotation=90);
```

## Example 6. Rotate the axes clockwise

Even after using the matplotlib.pyplot.xticks(rotation=90) function, if the category labels do not fit well, you can rotate the axes.

```python
# Plot the pokemon type on a horizontal bar chart

type_order = pokemon['type_1'].value_counts().index
```
