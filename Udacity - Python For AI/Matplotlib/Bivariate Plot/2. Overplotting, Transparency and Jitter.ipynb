## [Overplotting, Transparency and Jitter](https://youtu.be/BGqR-nxgMtg?si=GDphuiUHP2OOaCYS)
Overplotting refers to a situation where there is a large number of data points in a small area, making it difficult to see the actual relationship between the variables in a scatter plot. This can happen when there is a high density of points or when the data points are discrete.

To address overplotting, we can use techniques like sampling, transparency, and jitter. Sampling involves plotting a smaller number of points to make the patterns and relationships more visible. For example, if we plot 800 points instead of 10,000, we can see the gaps between the peaks more clearly.

Transparency is another technique where we plot all the data points but make each point slightly transparent. This reduces the brightness of individual points but allows us to see the overlapping regions when points overlap.

Jitter is a technique where we add a small random value to the position of each point. This helps to spread out the points and avoid overlapping, making it easier to see the density and distribution of the data.

These techniques can be used individually or in combination to improve the visualization of data and reveal meaningful patterns and relationships.

Certainly! Let's say we have a dataset where the x-axis represents the age of individuals and the y-axis represents their income. However, due to overplotting, it's difficult to see the individual data points clearly.

To address this, we can apply jitter to the data points. Jittering involves adding a small random noise to the position of each point. This helps to spread out the points that have the same values, creating a small variation in their positions.

For example, let's consider a scenario where we have three individuals with the same age of 30 and different income levels. Without jitter, these three points would overlap and appear as a single point on the plot. However, by applying jitter, we can slightly vary the position of each point, making them spread out within a small range around the age value of 30.

By adding jitter, we can now see three distinct points on the plot, representing the different income levels for individuals of the same age. This helps us identify and interpret trends or patterns in the data more effectively.


## Example 1. Jitter - Randomly add/subtract a small value to each data point

```python
# Import Necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# Read the CSV File 
fuel_econ = pd.read_csv('./fuel-econ.csv')
```

```python
plt.figure(figsize=[20, 5])

# Plot on Left Simple Scatter 
plt.subplot(1, 2, 1)
sb.regplot(data= fuel_econ, x = 'year', y = 'comb', truncate=True)

# Plot on Right- Simple Scatter
plt.subplot(1,2,2)
sb.regplot(data=fuel_econ, x='year', y='comb', truncate=False, x_jitter=0.3 )
```

In the left scatter plot above, the degree of variability in the data and strength of relationship are fairly unclear. In cases like this, we may want to employ transparency and jitter to make the scatterplot more informative. The right scatter plot has a jitter introduced to the data points.

## Example 2. Plot with both Jitter and Transparency

You can add transparency to either scatter() or regplot() by adding the "alpha" parameter set to a value between 0 (fully transparent, not visible) and 1 (fully opaque). See the example belo

```python
# Define Figure Size
plt.figure(figsize=[20, 5])

# Left Scatter Plot by seaborn libray 
plt.subplot(1,2,1)
sb.regplot(data=fuel_econ, x='year', y='comb', truncate=False, x_jitter=0.3, scatter_kws={'alpha': 1/20})

# Right Scatter Plot by matplotlib library
plt.subplot(1,2,2)
plt.scatter(data=fuel_econ, x='year', y='comb', alpha=1/20)
```

```python
# Plot in single Figure
sb.regplot(data=fuel_econ, x='year', y='comb', truncate=False, x_jitter=0.3, scatter_kws={'alpha': 1/20})
plt.scatter(data=fuel_econ, x='year', y='comb', alpha=1/20)
```

In the plot above, the jitter settings will cause each point to be plotted in a uniform 0.3 range of their true values. Note that transparency has been changed to be a dictionary assigned to the "scatter_kws" parameter.
