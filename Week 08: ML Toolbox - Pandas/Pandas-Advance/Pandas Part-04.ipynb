## Join, Merge & Apply method

```python
import pandas as pd
```

```python
data1 = pd.DataFrame({
    'Column1': [1, 2, 3, 4, 5],
    'Column2': [5, 6, 7, 8, 9],
    'Column3': [1, 3, 5, 7, 9]
    
})
```

```python
data1
```

```python
data2 = pd.DataFrame({
    'Column1':[1,2,45,6,67],
    'Column4':[56,5,6,7,8],
    'Column5':[3,56,5,6,6]
})
```

```python
data2
```

```python
#merge method

pd.merge(data1, data2) # একটা কলামকে কমন ধরে তার রো যে যে  ভ্যালুগুলো একই সেগুলো মার্জ করে 
```

```python
pd.merge(data1,data2,how = 'left') # প্যারা মিটারে দেওয়া প্রথম ডেটা ফ্রেমের  ভ্যালুগুলো রেখে দিবে, দ্বিতীয় ডেটা ফ্রেম এর মধ্যে যেগুলো কম ঐ রোগুলো বসাবে বাকিগুলো NaN বসাইয়া দিবে 
```

```python
pd.merge(data1,data2,how = 'right')
```

```python
pd.merge(data1,data2,how = 'outer')
```

```python
pd.merge(data1,data2,how = 'cross')
```

# Join 

### জয়েন মেথড ইনডেক্স ধরে কাজ করে, অন্যদিকে মার্জ মেথড কমন কলাম ধরে কাজ করে 

```python
data1 = pd.DataFrame({
    'Column1' : [1, 2,4, 5, 6],
    'Column2' : [4, 5, 6, 7, 8],
    'Column3' : [3, 4, 5, 6, 6]
}, index = ['a', 'b', 'c', 'd', 'e'])
```

```python
data2 = pd.DataFrame({
    'Column11' : [1, 2,4, 5, 6],
    'Column22' : [4, 5, 6, 7, 8],
    'Column33' : [3, 4, 5, 6, 6]
}, index = ['a', 'b', 'h', 'i', 'j'])
```

```python
data1
```

```python
data2
```

```python
data1.join(data2)
```

```python
data1.join(data2, how='right')
```

```python
data1.join(data2, how='inner')
```

```python
data1.join(data2, how='outer')
```

```python
data1.join(data2, how='cross')
```

## Apply 

apply method এ প্যারামিটার হিসেবে একটা ফাংশন নেয়। এরপর এক এক করে ডেটা ফ্রেম থেকে যে কলামের উপর apply method চালানো হয়েছে সেটা থেকে ডেটা নিবে এবং ফাংশনে পাস করবে। এরপর ফাংশন ঐটা ডেটা উপর অপারেশন করে দিবে। 

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.head()
```

```python
df['Fare_BDT'] = df['Fare'].apply(lambda x : x*100)
```

```python
df.head()
```

```python
df['name_len'] = df['Name'].apply(len)
```

```python
df
```

```python
def cat_fare(x):
    if x > 10:
        return 'Chip'
    elif x >= 10 and x < 20:
        return 'Mid'
    else:
        return 'High'
```

```python
df['cat_fare'] = df['Fare'].apply(cat_fare)
```
